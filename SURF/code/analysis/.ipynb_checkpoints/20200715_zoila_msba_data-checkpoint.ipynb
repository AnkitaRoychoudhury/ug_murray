# Zoila's MSBA Data Analysis
In this notebook, we will take a look at Zoila's experimental data. File is under data/2020_07_02_aHL_outside_MsbA<br>
7.15.2020

```python
# Import cell
import numpy as np
import pandas as pd
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
import colorcet
```

```python
df = pd.read_excel('../../data/zoila_2.xlsx')
df.head()
```

```python
df.index
```

```python
palette = colorcet.b_glasbey_category10
```

```python
p = bokeh.plotting.figure(width = 450, height = 400,
                         title = "Zoila Data (2020_07_02_aHL_outside_MsbA)",
                         x_axis_label = 'Time Index',
                         y_axis_label = 'Fluorescence')
p.line(df.index, df['G22'], color = palette[0], line_width = 2, legend_label = 'sfGFP-MsbA')
p.line(df.index, df['G23'], color = palette[1], line_width = 2, legend_label = 'MsbA-sfGFP')
p.line(df.index, df['G24'], color = palette[2], line_width = 2, legend_label = 'sfGFP')

x_line = [50,50]
y_line = [0,3500]
p.line(x_line, y_line, line_width = 2,alpha = 0.5, color= palette[4])



bokeh.io.show(p)
```

We just want to be able to estimate the time at which stuff tends to come to steady state in the membrane. Seems between 2e7 and 3e7. With new x-axis, seems to be at 50.
