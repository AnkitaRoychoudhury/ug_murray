```python
import pandas as pd

import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
from bokeh.layouts import row

import colorcet
```

## Look at Isobutanol without Overlay
---

```python
df_iso = pd.read_excel('../../../private/bowie_lab_2020/isobutanol_data.xlsx', header = 1)
df_iso.head()
```

```python
# Convert g/L to mM
df_iso['Concentration (mM)'] = df_iso['Titer (g/L)'] / 74.122 * 1000
```

```python
df_iso
```

```python
len(df_iso)
```

```python
df_iso_over = pd.read_excel('../../../private/bowie_lab_2020/isobutanol_data_2.xlsx', header = 1)
df_iso_over.head()
```

```python
# Convert g/L to mM
```

```python
colors = colorcet.b_glasbey_category10
```

```python
p = bokeh.plotting.figure(width = 370,
                         height = 350, title = 'Isobutanol Bowie Lab 2 Data',
                         x_axis_label = 'Time (hrs)',
                         y_axis_label = 'Titer (g/L)')

p.line(df_iso['Time (h)'], df_iso['Titer (g/L)'], line_width = 2, color = colors[0], legend_label = 'without overlay')
p.line(df_iso_over['Time (h)'], df_iso_over['Titer (g/L)'], line_width = 2, color = colors[1], legend_label = 'with overlay')

p.legend.location = 'bottom_right'

# Add error bars
for i in range(len(df_iso)):
    p.line([df_iso['Time (h)'][i], df_iso['Time (h)'][i]], [df_iso['Titer (g/L)'][i] - df_iso['SD'][i], df_iso['Titer (g/L)'][i] + df_iso['SD'][i]], color = colors[0])
    
for j in range(len(df_iso_over)):
    p.line([df_iso_over['Time (h)'][j], df_iso_over['Time (h)'][j]], [df_iso_over['Titer (g/L)'][j] - df_iso_over['SD'][j],
                                                                      df_iso_over['Titer (g/L)'][j] + df_iso_over['SD'][j]], color = colors[1])  

bokeh.io.show(p)
```

## Look at NADPH data
---

```python
df_nadph = pd.read_excel('../../../private/bowie_lab_2020/nadph_data.xlsx', header = 1, keep_default_na = False)
df_nadph = df_nadph[:10]
df_nadph.head()
```

```python
df_nadph_over = pd.read_excel('../../../private/bowie_lab_2020/nadph_data_2.xlsx', header = 1, keep_default_na = False)
# df_nadph = df_nadph[:10]
df_nadph_over.head()
```

```python
p1 = bokeh.plotting.figure(width = 370,
                         height = 350, title = 'Metabolite Bowie Lab 2 Data',
                         x_axis_label = 'Time (hrs)',
                         y_axis_label = 'Concentration (mM)')

p1.line(df_nadph['Time (h)'], df_nadph['mM'], line_width = 2, color = colors[0], legend_label = 'nadph without overlay')
p1.line(df_nadph_over['Time (h)'], df_nadph_over['mM'], line_width = 2, color = colors[1], legend_label = 'nadph with overlay')
p1.circle(88, 2.365523641, line_width = 2, color = colors[2],size =4, legend_label = 'atp without overlay')
p1.circle(88, 5.279271563, line_width = 2, color = colors[3],size =4, legend_label = 'atp with overlay')




p1.legend.location = 'top_center'

# Add error bars
for i in range(len(df_nadph)):
    p1.line([df_nadph['Time (h)'][i], df_nadph['Time (h)'][i]], [df_nadph['mM'][i] - df_nadph['SD'][i], df_nadph['mM'][i] + df_nadph['SD'][i]], color = colors[0])
    
for j in range(len(df_nadph_over)):
    p1.line([df_nadph_over['Time (h)'][j], df_nadph_over['Time (h)'][j]], [df_nadph_over['mM'][j] - df_nadph_over['SD'][j],
                                                                      df_nadph_over['mM'][j] + df_nadph_over['SD'][j]], color = colors[1])  
p1.line([88,88], [2.365523641-0.418811263,2.365523641+0.418811263] , line_width = 2, color = colors[2], legend_label = 'atp without overlay')
p1.line([88,88], [5.279271563-1.086223609,5.279271563+1.086223609] , line_width = 2, color = colors[3], legend_label = 'atp with overlay')



bokeh.io.show(p1)
```

Overlay was included so that isobutanol will partition preferentially to the organic layer, keeping concentrations low enough in aqueous phase to keep thermodynamic gradient for more isobutanol production (TAKEN FROM BOWIE LAB PAPER DRAFT)<br>
We only need to consider the ones without overlay.
