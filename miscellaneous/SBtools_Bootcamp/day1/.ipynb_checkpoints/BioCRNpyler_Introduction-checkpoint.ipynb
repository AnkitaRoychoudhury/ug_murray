## What is a CRN?
A CRN is a widely established model of chemistry and biochemistry.
* A set of species $S$
* A set of reactions $R$ interconvert species $I_r$ to $O_r$

\begin{align}
\\
I \xrightarrow[]{\rho(s)} O
\\
\end{align}

 * $I$ and $O$ are multisets of species $S$. 
 * $\rho(s): S \to \mathbb{R}$ is a function that determines how fast the reaction occurs.



## CRNs in BioCRNpyler:

To create CRN models in BioCRNpyler, you need two kinds of objects:

* `Species` : Used to represent each individual species in your model. 

* `Reaction`: Used to represent each reaction that refers to `Species` objects and has propensity options.

Syntax:
```
species_object = Species(name = "name_string")

list_of_inputs = [species_object]

list_of_outputs = []

reaction_object = Reaction(list_of_inputs, list_of_outputs, k = rate_parameter_value)
```



### Example
\begin{equation}
\emptyset  \underset{1}{\overset{100}\rightleftharpoons}  2 A
\end{equation}


We have two separate reactions (one reversible reaction), each with a mass-action propensity rate parameter.

### Using BioCRNpyler to model this CRN:

````python
%matplotlib inline
%config InlineBackend.figure_format = "retina"
from matplotlib import rcParams
rcParams["savefig.dpi"] = 100
rcParams["figure.dpi"] = 100
rcParams["font.size"] = 20
````

````python
import numpy as np
import pylab as plt

from biocrnpyler.chemical_reaction_network import Species, Reaction, ChemicalReactionNetwork
#Create a CRN

# Create all your species:
A = Species(name = "A")

# Create CRN model:
# Note the usage of stoichiometry coefficients > 1:

CRN = ChemicalReactionNetwork(species = [A], 
                              reactions = [Reaction([], [A, A], k = 100),
                                           Reaction([A, A], [], k = 1)])
````

````python
#Species, reactions, and CRNs can all be directly printed
print("\n CRN Model:\n", CRN)
timepoints = np.linspace(0, 0.3, 60)
plt.figure()
R = CRN.simulate_with_bioscrape(timepoints = timepoints, initial_condition_dict = {"A":0})
plt.plot(timepoints, R["A"], color = 'black', lw = 4)
# plt.title('Concentration Trajectory Dynamics')
plt.xlabel("Time")
plt.ylabel("Concentration");
plt.show()
````

# Helpful tricks and utility functions:

## 1. Adding material_type to Species:

````python
A = Species(name = "A", material_type = "DNA")
B = Species(name = "B", material_type = "protein")

print('The species are:')
print(A)
print(B)

print('\nNote that species names cannot have semicolons, colons, or other special characters.' +
      'Underscores are allowed. First character must always be an alphabet.')
````

## 2. Adding a reversible reactions in one-line:

````python
reaction1 = Reaction([A], [B, B], k = 100, k_rev = 10) #type always defaults to massaction
reaction2 = Reaction([B, B], [], k = 20) #type always defaults to massaction

print('The reactions are:')
print(reaction1)
print(reaction2)
````

````python
print("Writing the CRN model as an SBML file (to output, exchange, standardize the model)...")
CRN.write_sbml_file('simple_CRN.xml');
print("Done!")
````

##  3. BioCRNpyler outputs SBML models: What is SBML? 

### SBML is a standardized language to exchange models : sbml.org/



## 4. Simulation options: 

1. Bioscrape directly : Use `simulate_with_bioscrape` for a CRN object.
2. Bioscrape via SBML: Use `simulate_with_bioscrape_via_sbml` for a CRN object. Needs an SBML model file (a .XML file).
3. Any other simulator: BioCRNpyler outputs SBML. SBML is a standard language that hundreds of software tools support. For example:
    * COPASI
    * RoadRunner
    * MATLAB SimBiology
