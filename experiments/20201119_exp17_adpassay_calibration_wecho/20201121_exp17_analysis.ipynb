```python
import bokeh
import bokeh.plotting
bokeh.io.output_notebook()
from bokeh.layouts import Row

import numpy as np
import scipy
import scipy.stats
```

```python
vals = [79,41,9,14,90,11,46,21,15,7,38,108,
        11,97,27,31,14,28,68,11,46,
        8,10,9,6,13,164]
vals_nm = [val*10**6 for val in vals]
wells = ['A1', 'A2', 'A3', 'A4', 'A5', 'A6', 'A7', 'A8', 'A9', 'A10', 'A11', 'A12',
        'B1', 'B2', 'B3', 'B4', 'B5', 'B6', 'B8', 'B10', 'B12',
        'C1', 'C2', 'C3', 'C4', 'C5', 'C12']


descriptive_wells = []

adp_atp_concs = [1000, 800, 600, 400, 200, 100, 50, 40, 30, 20, 10,0.000000001 ,
                100, 80, 60, 40, 20, 10, 4, 2,0.000000001,
                10, 8, 6, 4, 2,  0.000000001]
```

```python
p_log = bokeh.plotting.figure(width = 450, height = 300,
                          title = 'Log ADP Assay Calibration',
                         x_axis_label = 'Log ADP Concs (nM)',
                         y_axis_label = 'Log Lum. Values',
                         x_axis_type = 'log',
                         y_axis_type = 'log')
p_log.circle(adp_atp_concs, vals_nm)


p = bokeh.plotting.figure(width = 450, height = 300,
                          title = 'ADP Assay Calibration',
                         x_axis_label = 'ADP Concs (nM)',
                         y_axis_label = 'Lum. Values',)
p.circle(adp_atp_concs, vals_nm)

bokeh.io.show(Row(p_log, p))
```

```python
par = np.polyfit(np.log(vals_nm), np.log(adp_atp_concs), 1)
```

```python
slope = par[0]
intercept = par[1]
```

```python
x = np.log(adp_atp_concs)
x = np.linspace(0.0000001,10**7, 1000)
y_predicted =[-slope * i + intercept for i in x]
```

```python
tooltips = [
    ("ATP Conc", "@ATP_Conc"),
    ("Values", "@Values"),
]
p = bokeh.plotting.figure(width = 450, height = 300,
                         x_axis_label = 'ATP (nM)',
                         y_axis_label = 'Fluorescence',
                          x_axis_type = 'log',
                          y_axis_type = 'log',
                         title = 'ATP Assay Calibration (Gain 135)',
                         tooltips = tooltips)
#p.circle(np.log([a / (1e12) for a in atp]), np.log(vals))
p.circle(adp_atp_concs, vals_nm)

p.line(x, y_predicted, color= 'orange', line_width = 2)
#p.circle(x,y_predicted, color = 'orange')

bokeh.io.show(p)
```
