```python
import numpy as np
import numba

import bokeh.io
import iqplot
bokeh.io.output_notebook()

@numba.njit
def time_random_walk():
    # Set parameters
    # you take either -1 step or +1 step (coin flip)
    prob = 0.5

    # use for exponential distribution (average time = 0.5 s)
    beta = 2
    
    # set position x=0, steps=0
    position = 0
    i=0
    
    # randomly move +1 or -1, loop stops when x=1
    while position != 1:
        i += 1
        test = np.random.uniform(0,1)
        if test >= prob:
            position += 1
        else:
            position -= 1
    
    # draw values out of exponential distribution
    N_steps = i           
    exp_draw = np.random.exponential(beta, size=N_steps)
    total_time = np.sum(exp_draw)
    return total_time

time_random_walk()

def time_list(n):
    time_list_func=[time_random_walk() for _ in range(n)]
    return time_list_func

time_list_10000 = time_list(10000)

time_list_array = np.array(time_list_10000)
p = iqplot.ecdf(
    data=time_list_array,
    x_axis_type='log',
   # x_range=(0,time_list_array.max())
)

bokeh.io.show(p)
```
